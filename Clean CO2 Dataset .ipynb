```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#import dataset 
raw_co2= pd.read_csv('owid-co2-data.csv')
raw_co2
```

```python
#Drop columns 
raw_co2.drop(columns=['iso_code','co2_growth_prct','co2_growth_abs','consumption_co2_per_capita','share_global_co2',
                      'cumulative_co2','share_global_cumulative_co2','consumption_co2_per_gdp','co2_per_gdp',
                      'co2_per_unit_energy','cement_co2_per_capita','coal_co2_per_capita',
                      'flaring_co2_per_capita','gas_co2_per_capita','oil_co2_per_capita',
                      'other_co2_per_capita','trade_co2_share','share_global_cement_co2',
                      'share_global_coal_co2','share_global_flaring_co2','share_global_gas_co2',
                      'share_global_oil_co2','share_global_other_co2','cumulative_cement_co2',
                      'cumulative_coal_co2','cumulative_flaring_co2','cumulative_gas_co2',
                      'cumulative_oil_co2','cumulative_other_co2','share_global_cumulative_cement_co2',
                      'share_global_cumulative_coal_co2','share_global_cumulative_flaring_co2',
                      'share_global_cumulative_gas_co2','share_global_cumulative_oil_co2',
                      'share_global_cumulative_other_co2','methane_per_capita',
                      'nitrous_oxide_per_capita','energy_per_capita','primary_energy_consumption',
                      'energy_per_gdp'], inplace = True)
raw_co2
```

```python
#define values
values = range(1700,1850)

#drop rows that contain any value in the list
raw_co2 = raw_co2[raw_co2.year.isin(values) == False]
raw_co2
```

```python
raw_co2 = raw_co2.dropna(subset=['co2'])
raw_co2 = raw_co2.reset_index(drop=True)
raw_co2
```

```python
raw_co2.index += 1
raw_co2
```

```python
# save new dataset as a excel sheet 
raw_co2.to_excel("updatedco2.xlsx",index=False)
```
